## Scraping Artist's Data and Track's Audio Features~

```python
nama_band = []
genre_band = []
album = []
album_tipe = []
album_tgl_rilis = []
judul_track = []
populer_track = []
fenergy = []
fdanceability = []
ftempo = []
finstrumentalness = []
floudness = []
fspeech = []
fvalence = []
fliveness = []
fdurasi = []
```

```python
# Start Scraping!

for artist_URI in artists_URI:
    # Fetch artist information once, store it in artist_info
    artist_info = sp.artist(artist_URI)

    # Get general band data
    band_name = artist_info["name"]
    band_genre = artist_info["genres"]

    # Get all albums of the artist
    artist_albums = sp.artist_albums(artist_URI)["items"]

    # Iterate through the albums
    for album_info in artist_albums:
        album_name = album_info["name"]
        album_type = album_info["album_type"]
        album_release_date = album_info["release_date"]
        album_total_tracks = album_info["total_tracks"]
        album_id = album_info["id"]  # Fetch the album ID to get the tracks

        # Get tracks for each album
        track_info = sp.album_tracks(album_id)["items"]

        # Iterate through the tracks in the album
        for track in track_info:
            track_name = track["name"]
            track_id = track["id"]

            # Get track features (danceability, etc.)
            track_features = sp.audio_features(track_id)
            track_details = sp.track(track_id)
            track_pop = track_details["popularity"]

            # Check if track_features is valid (API might return None if not available)
            if track_features and track_features[0]:
                track_danceability = track_features[0]['danceability']
                track_energy = track_features[0]['energy']
                track_inst = track_features[0]['instrumentalness']
                track_loud = track_features[0]['loudness']
                track_speech = track_features[0]['speechiness']
                track_tempo = track_features[0]['tempo']
                track_liveness = track_features[0]['liveness']
                track_valence = track_features[0]['valence']
                track_duration = track_features[0]['duration_ms']

                nama_band.append(band_name)
                genre_band.append(band_genre)
                album.append(album_name)
                album_tgl_rilis.append(album_release_date)
                album_tipe.append(album_type)
                judul_track.append(track_name)
                fenergy.append(track_energy)
                fdanceability.append(track_danceability)
                ftempo.append(track_tempo)
                finstrumentalness.append(track_inst)
                floudness.append(track_loud)
                fspeech.append(track_speech)
                fliveness.append(track_liveness)
                fvalence.append(track_valence)
                populer_track.append(track_pop)
                fdurasi.append(track_duration)


                # Print track information
               # print(f"Track: {track_name} | Danceability: {track_danceability}")
            else:
                print(f"Track: {track_name} | No audio features available.")

        # Print album information
        #print(f"Data telah terinput")
        #print()

```
